```python
x = df[['comment']]
y = df.sentiment
```

```python
x.shape
y.shape
```

```python
import jieba
def word_cut(text):
    return " ".join(jieba.cut(text))
```

```python
x['cutted_comment'] = x.comment.apply(word_cut)
```

```python
x.cutted_comment[:5]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=None, shuffle=True)
```

```python
def get_custom_stopwords(stop_words_file):
    with open(stop_words_file) as f:
        stopwords = f.read()
    stopwords_list = stopwords.split('\n')
    custom_stopwords_list = [i for i in stopwords_list]
    return custom_stopwords_list
```

```python
stop_words_file = "stopwordsHIT.txt"
stopwords = get_custom_stopwords(stop_words_file)
```

```python
stopwords[-10:]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(max_df = 0.8,
                       min_df = 3,
                       token_pattern= u'(?u)\\b[^\\d\\W]\\w+\\b',#u'[\u4e00-\u9fa5]+'
                       stop_words=frozenset(stopwords),
                       ngram_range=(1,1))
```

```python
term_matrix = pd.DataFrame(vec.fit_transform(x_train.cutted_comment).toarray(), columns=vec.get_feature_names())
```

```python
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB(alpha=1.0, fit_prior=True, class_prior=None)
```

```python
from sklearn.pipeline import make_pipeline
pipe = make_pipeline(vec, nb)
```
