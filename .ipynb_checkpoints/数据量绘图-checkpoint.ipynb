```python
import matplotlib.pyplot as plt#加载可视化绘图工具
```

```python
range = [2,4,6,8,10,12]
scores_vec =[0.8989,0.90075,0.9021,0.9021,0.9023,0.9031]
```

```python
scores_tfidf = [0.8596,0.8646,0.864,0.8643,0.8651,0.8665] 
```

```python
plt.suptitle('CountVectorizer')
plt.plot(range,scores_vec)
plt.xlabel('Date')
plt.ylabel('Accuracy')
plt.show()
```

```python
plt.suptitle('TfidfVectorizer')
plt.plot(range,scores_tfidf)
plt.xlabel('Date')
plt.ylabel('Accuracy')
plt.show()
```
