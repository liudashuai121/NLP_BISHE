```python
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(max_df = 0.8,
                       min_df = 3,
                       token_pattern= u'[\u4e00-\u9fa5]+',
                       #stop_words=frozenset(stopwords),
                       ngram_range=(1,1))
```

```python
documents = ['张云雷 希望 大家 在 追星 的 同时 保持 理智',
            '关注 作品 远离 生活',
            '支持 保护 各位 角儿',
            '相声 不是 追星 捧 角儿 得 有度',
            '张云雷 这么 做 真的 好吗 抵制']
count_vec = CountVectorizer()

count_data = count_vec.fit_transform(documents)
```

```python
term_matrix1 = pd.DataFrame(count_vec.fit_transform(documents).toarray(), columns=count_vec.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vec = TfidfVectorizer()
count_data = tfidf_vec.fit_transform(documents)
```

```python
term_matrix2 = pd.DataFrame(tfidf_vec.fit_transform(documents).toarray(), columns=tfidf_vec.get_feature_names())
```
